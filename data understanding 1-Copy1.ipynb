```python
# ============================================
# 4.1.3 Variable Selection â€“ Statistical Relevance
# Correlation Analysis for Engineering Effort
# ============================================

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# --- 1. Load your main working dataset ---
# IMPORTANT: replace with your actual file
df = pd.read_csv("daskan_master_dataset.csv")

# --- 2. Identify numerical variables for correlation ---
numeric_cols = df.select_dtypes(include=[np.number]).columns.tolist()

print("Numerical columns considered for correlation:")
print(numeric_cols)

# --- 3. Compute correlation matrix ---
corr_matrix = df[numeric_cols].corr()

# Display correlation with target variable (e.g., total_project_hours)
target = "total_project_hours"   # Change if your target column name differs

if target in numeric_cols:
    corr_target = corr_matrix[target].sort_values(ascending=False)
    print("\nCorrelation of all numerical variables with", target)
    print(corr_target)
else:
    print(f"\nWarning: {target} is not in your numeric columns.")

# --- 4. Plot correlation heatmap ---
plt.figure(figsize=(12, 10))
sns.heatmap(corr_matrix, annot=False, cmap="coolwarm")
plt.title("Correlation Heatmap (Numerical Variables)")
plt.show()

# --- 5. Identify low-correlation variables (optional) ---
# Threshold can be changed
threshold = 0.10  
low_corr = corr_target[abs(corr_target) < threshold]

print("\nVariables with low correlation to engineering effort (|corr| < 0.10):")
print(low_corr)

# --- 6. Identify high-correlation variables (optional for feature selection) ---
high_corr = corr_target[abs(corr_target) >= threshold]

print("\nVariables with meaningful correlation to engineering effort:")
print(high_corr)
```
